```python
import pandas as pd

# 1. Load Data
df = pd.read_csv('MegaDepth1500_summary.csv')

# 2. Tentukan kolom score (Select semua kolom KECUALI Matcher dan Time)
#    Kita mengecualikan 'Matcher', 'Time Min', 'Time Max', dan 'Time Avg'
cols_to_exclude = ['Matcher', 'Min Time', 'Max Time', 'Avg Time']
score_cols = [c for c in df.columns if c not in cols_to_exclude]

# 3. Hitung rata-rata score per baris (per matcher)
df['Avg Score'] = df[score_cols].mean(axis=1)

# 4. Ambil Top 10 berdasarkan Time Avg terendah
top10_speed_score = df.nsmallest(20, 'Avg Time')

# 5. Tampilkan hasilnya
#    Format float agar rapi (2 desimal)
pd.options.display.float_format = '{:,.2f}'.format

print("="*50)
print("Top 20 Time Avg (Lowest) with Overall Avg Score MegaDepth1500")
print("="*50)
print(top10_speed_score[['Matcher', 'Avg Time', 'Avg Score']].to_string(index=False))
```
