```python
# Daftar kolom yang ingin diambil top 10-nya
# Kolom Time akan diurutkan ascending (terendah), kolom lainnya descending (tertinggi)

# Contoh untuk beberapa kolom:
metrics = ['Total AUC@1', 'Total AUC@3', 'Total MHA@3', 'Time Avg']

# Dictionary untuk menyimpan hasil top 10 per kolom
top_10_results = {}

for metric in metrics:
    if 'Time' in metric:
        # Untuk kolom Time, ambil 10 nilai TERENDAH
        top_10_results[metric] = df.nsmallest(10, metric)[['Matcher', metric]]
    else:
        # Untuk kolom lainnya, ambil 10 nilai TERTINGGI
        top_10_results[metric] = df.nlargest(10, metric)[['Matcher', metric]]

# Menampilkan hasil
for metric, data in top_10_results.items():
    print(f"\n{'='*60}")
    print(f"Top 10 untuk {metric}:")
    print('='*60)
    print(data.to_string(index=False))
```

```python
import pandas as pd

# 1. Load Data
df = pd.read_csv('Hpatch_summary.csv')

# 2. Ambil Top 10 berdasarkan Time Avg terendah (Ascending)
top10_time = df.nsmallest(10, 'Time Avg')

# 3. Definisikan kolom untuk masing-masing tabel
# Kita selalu sertakan 'Matcher' dan 'Time Avg' agar konteksnya jelas
base_cols = ['Matcher', 'Time Avg']
total_cols = [c for c in df.columns if c.startswith('Total')]
illum_cols = [c for c in df.columns if c.startswith('Illumination')]
view_cols = [c for c in df.columns if c.startswith('Viewpoint')]

# 4. Tampilkan Tabel 1: Total Metrics
print("="*100)
print("Top 10 Time Avg (Lowest) - Table 1: TOTAL Metrics")
print("="*100)
print(top10_time[base_cols + total_cols].to_string(index=False))
print("\n")

# 5. Tampilkan Tabel 2: Illumination Metrics
print("="*100)
print("Top 10 Time Avg (Lowest) - Table 2: ILLUMINATION Metrics")
print("="*100)
print(top10_time[base_cols + illum_cols].to_string(index=False))
print("\n")

# 6. Tampilkan Tabel 3: Viewpoint Metrics
print("="*100)
print("Top 10 Time Avg (Lowest) - Table 3: VIEWPOINT Metrics")
print("="*100)
print(top10_time[base_cols + view_cols].to_string(index=False))
```

```python
import pandas as pd

# 1. Load Data
df = pd.read_csv('Hpatch_summary.csv')

# 2. Tentukan kolom score (Select semua kolom KECUALI Matcher dan Time)
#    Kita mengecualikan 'Matcher', 'Time Min', 'Time Max', dan 'Time Avg'
cols_to_exclude = ['Matcher', 'Time Min', 'Time Max', 'Time Avg']
score_cols = [c for c in df.columns if c not in cols_to_exclude]

# 3. Hitung rata-rata score per baris (per matcher)
df['Avg Score'] = df[score_cols].mean(axis=1)

# 4. Ambil Top 10 berdasarkan Time Avg terendah
top10_speed_score = df.nsmallest(20, 'Time Avg')

# 5. Tampilkan hasilnya
#    Format float agar rapi (2 desimal)
pd.options.display.float_format = '{:,.2f}'.format

print("="*50)
print("Top 20 Time Avg (Lowest) with Overall Avg Score HPatch")
print("="*50)
print(top10_speed_score[['Matcher', 'Time Avg', 'Avg Score']].to_string(index=False))
```
